```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img =np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='VARNI',org=(90,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    plt.show()
```

```python
img = load_img()
```

```python
display_img(img)
```

## Erosion

```python
kernel = np.ones((5, 5), np.uint8)
```

```python
Erosion=cv2.erode(img, kernel, iterations=1)
```

```python
display_img(Erosion)
```

## Dilation

```python
kernel = np.ones((5, 5), np.uint8)
```

```python
dilation=cv2.dilate(img, kernel, iterations=1)
```

```python
display_img(dilation)
```
